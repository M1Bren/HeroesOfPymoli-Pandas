```python
# Dependencies and Setup
import pandas as pd
# pd.options.mode.chained_assignment = None  # default='warn

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
# Sample DataFrame of the purchase data based on each Purchase ID (where username/SN and Item Name can repeat)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Get a numerical count of total players (no repeats)
totalplayers = purchase_data["SN"].nunique()
totalplayers_dict = [{"Total Players": totalplayers}]

# Put this number into a new DataFrame
tpdf = pd.DataFrame(totalplayers_dict)
tpdf
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Make the DataFrame of statistics
avg_price = round(purchase_data["Price"].mean(),2)
uniq_item_num = purchase_data["Item Name"].nunique()
purchase_num = purchase_data["Purchase ID"].nunique()
total_rev = round(purchase_data["Price"].sum(),2)
Purchase_analysis = pd.DataFrame([{"Number of Unique Items": uniq_item_num,
                                   "Average Price": avg_price,
                                   "Number of Purchases": purchase_num,
                                   "Total Revenue": total_rev}])

#Add formatting using the mapping format
Purchase_analysis["Average Price"] = Purchase_analysis["Average Price"].map("${:.2f}".format)
Purchase_analysis["Total Revenue"] = Purchase_analysis["Total Revenue"].map("${:,}".format)
Purchase_analysis
```

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Drop duplicates on purchase_data file.  Keep the first entry of the username that appears in SN
purdata_uniq = purchase_data.drop_duplicates(subset='SN', keep='first', inplace=False)

# Change 'Purchase ID' column to 'Total Count' 
gender_df = purdata_uniq.rename(columns={'Purchase ID':'Total Count'})

# GroupBy based on gender and set it to count by default.
# Note that 'Percentage Column' will later show its own values
gender_df = gender_df.groupby(['Gender'])
gender_df = gender_df.count()

# Come up with stats for 'Percentage of Players' column
female_percent = round((gender_df.iloc[0, 1]) / totalplayers * 100, 2)
male_percent = round((gender_df.iloc[1, 1]) / totalplayers * 100, 2)
other_percent = round((gender_df.iloc[2, 1]) / totalplayers * 100, 2)

# Add this new column to gender_df and select for it and 'Total Count'
gender_df["Percentage of Players"] = [female_percent, male_percent, other_percent]
gender_df = gender_df[["Total Count", "Percentage of Players"]]
gender_df["Percentage of Players"] = gender_df["Percentage of Players"].map("{:}%".format)

gender_df
```

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Calculate data columns for final groupby dataframe
tot_pur = purchase_data.groupby(['Gender']).sum()['Price'].rename('Total Purchase Value')
avg_pur = purchase_data.groupby(['Gender']).mean()['Price'].rename('Average Purchase Price')
pur_cnt = purchase_data.groupby(['Gender']).count()['Price'].rename('Purchase Count')

# Calculate the average total purchase per person using purdata_uniq from previous cell
pur_per = tot_pur / purdata_uniq['Gender'].value_counts()

# Make a new groupby dataframe with the columns just created
# Since created columns were alll groupby'ed the same column (Gender), gender2_df will be a groupby as well
gender2_df = pd.DataFrame({'Purchase Count': pur_cnt, 
                           'Average Purchase Price':avg_pur,
                           'Total Purchase Value':tot_pur,
                           'Average Total Purchase per Person':pur_per})

# Data munging to improve format
gender2_df['Average Purchase Price'] = gender2_df['Average Purchase Price'].map('${:,.2F}'.format)
gender2_df['Total Purchase Value'] = gender2_df['Total Purchase Value'].map('${:,.2F}'.format)
# backspace continues this line on a new line
gender2_df['Average Total Purchase per Person'] = \
gender2_df['Average Total Purchase per Person'].map('${:,.2F}'.format)

# Display the aggregated groupby dataframe
gender2_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Estbalish bins according to age
age_bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 999]
bin_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

# Categorize players based on age bins and add this column to purdata_uniq
purdata_uniq['Age Ranges'] = pd.cut(purdata_uniq['Age'], age_bins, labels=bin_names)

# Calculate total count and percentage of players
tot_cnt = purdata_uniq['Age Ranges'].value_counts()
# Dividing a column of values by a number is totally valid
pct_play = tot_cnt / totalplayers * 100

# Make a new dataframe with above columns
# I guess making a dataframe with binned columns also gives a pivot table?
age_demogr = pd.DataFrame({'Total Count': tot_cnt, 'Percentage of Players': pct_play})

# Data munging to improve format
age_demogr['Percentage of Players'] = age_demogr['Percentage of Players'].map('{:.2F}%'.format)
age_demogr = age_demogr.sort_index()

# Display the DataFrame
age_demogr
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Make a new column in purchase_data named 'Age Ranges' using .cut()
purchase_data['Age Ranges'] = pd.cut(purchase_data['Age'], age_bins, labels=bin_names)


# Calculations for the columns achieved through groupby 'Age Ranges'
pur_cnt = purchase_data.groupby(['Age Ranges']).count()['Purchase ID'].rename('Purchase Count')
avg_price = purchase_data.groupby(['Age Ranges']).mean()['Price'].rename('Average Purchase Price')
tot_pur = purchase_data.groupby(['Age Ranges']).sum()['Price'].rename('Total Purchase Value')
# Average Total Purchase per Person will require purdata_uniq to avoid repeats of usernames!
avg_p_per = tot_pur / purdata_uniq.groupby(['Age Ranges']).count()['Purchase ID']

# Make the new DataFrame with the above columns.  Automatically grouped 
pur_anlys = pd.DataFrame({'Purchase Count': pur_cnt,
                          'Average Purchase Price': avg_price,
                          'Total Purchase Value': tot_pur,
                          'Average Total Purchase per Person': avg_p_per})
pur_anlys['Average Purchase Price']= pur_anlys['Average Purchase Price'].map('${:,.2F}'.format)
pur_anlys['Total Purchase Value']= pur_anlys['Total Purchase Value'].map('${:,.2F}'.format)
pur_anlys['Average Total Purchase per Person']= pur_anlys['Average Total Purchase per Person'].map('${:,.2F}'.format)

# Display the DataFrame
pur_anlys
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Calculations for the columns achieved through groupby 'SN'
pur_cnt = purchase_data.groupby('SN').count()['Purchase ID']
avg_pur = purchase_data.groupby('SN').mean()['Price']
tot_pur = purchase_data.groupby('SN').sum()['Price']

# Create a new DataFrame with Top Spenders dataset
top_spenders = pd.DataFrame({'Purchase Count': pur_cnt,
                             'Average Purchase Price': avg_pur,
                             'Total Purchase Value': tot_pur})

# 'Top Spenders' formatting
top_spenders['Average Purchase Price'] = top_spenders['Average Purchase Price'].map('${:,.2F}'.format)
top_spenders['Total Purchase Value'] = top_spenders['Total Purchase Value'].map('${:,.2F}'.format)
top_spenders = top_spenders.sort_values('Purchase Count', ascending=False)

# Display the 'Top Spenders' Table
top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a DataFrame of 'Most Popular Items' using the Merge Function of Pandas
tot_pur = purchase_data.groupby('Item Name').sum().reset_index()
item_price = purchase_data.groupby('Item ID').sum().reset_index()
pur_cnt = purchase_data.groupby('Item Name').count().reset_index()

# merge the above DataFrames to create columns by sum (_x) and count (_y)
mergeone = pd.merge(tot_pur, item_price, on='Price')
mergetwo = pd.merge(tot_pur, pur_cnt, on='Item Name')

# Make an 'Item Price' column
mergetwo['Item Price'] = (mergetwo['Price_x'] / mergetwo['Item ID_y'])

# Rename necessary columns and delete unnecessary ones
most_pop = mergetwo.rename(columns={'Purchase ID_y': 'Purchase Count',
                            'Price_x': 'Total Purchase Value'})
# Note that two brackets [[]] are necessary around the column headers
most_pop = most_pop[['Item Name', 'Purchase Count', 'Item Price', 'Total Purchase Value']]
most_pop = most_pop.sort_values('Purchase Count', ascending=False).head()

# Data Formatting
most_pop['Item Price'] = most_pop['Item Price'].map('${:,.2F}'.format)
most_pop['Total Purchase Value'] = most_pop['Total Purchase Value'].map('${:,.2F}'.format)

#most_pop = most_pop.style.format({'Item Price': '${:,.2F}','Total Purchase Value': '${,.2F}'})
most_pop.head()
```
